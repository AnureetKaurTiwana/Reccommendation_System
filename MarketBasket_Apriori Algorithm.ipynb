```python
## Main data is from UK, lets keep one country data for analysis
df=df[df.Country=='United Kingdom']
df.Country.value_counts()
```

```python
data=df.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().fillna(0)
data.head()
```

```python
data=(data>0).astype('int')
```

```python
data=data.astype('bool')
```

```python
### 22038 invoices and 4202 items
```

```python
## Apriori algorithm
```

```python
pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
## min_support 2.28 % means the item should be present in atleast 3% of the dataset
frequent_itemsets_plus=apriori(data,min_support=0.02,use_colnames=True).sort_values('support',ascending=False).reset_index(drop=True)
```

```python
frequent_itemsets_plus['len']=frequent_itemsets_plus['itemsets'].apply(lambda x:len(x))
```

```python
frequent_itemsets_plus
```

```python
frequent_itemsets_plus[frequent_itemsets_plus['len']>2]
```

```python
## what two items we can buy together or reccommend
from mlxtend.frequent_patterns import association_rules
```

```python
data.shape
```

```python
ass_rules=association_rules(frequent_itemsets_plus,metric='lift',min_threshold=1,num_itemsets=22038).sort_values('lift',ascending=False).reset_index(drop=True)
```
